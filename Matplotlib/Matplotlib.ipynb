```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Line Chat

```python
x = [1,2,3,4,5,6,7]
y = [30,31,24,29,23,39,35]
```

```python
plt.xlabel('Date')
plt.ylabel('Temprature')
plt.title('Weather')
plt.plot(x,y,color='green',linewidth = 5, linestyle = 'dotted')
```

```python
plt.plot(x,y,'rD')
```

```python
plt.plot(x,y,alpha=0.1)
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.ylabel('Temprature')
plt.xlabel('Days')
plt.title('Weather')
plt.plot(days,max_t, label='Max Temp')
plt.plot(days,min_t, label='Min Temp')
plt.plot(avg_t, label='Average Temp')
plt.legend(loc='best', shadow = True, fontsize='large')
plt.grid()
```

```python
plt.xticks(ypos,company)
plt.xlabel('Company Name')
plt.ylabel('US Dollar')
plt.title('Tech Stocks')
plt.bar(ypos+0.2,revenue,width=0.4, label = 'Revenue')
plt.bar(ypos-0.2,profit,width=0.4,label='Profit' )

plt.legend()
```
